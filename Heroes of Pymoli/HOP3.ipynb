## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9,50]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
age_data = purchase_data
age_data["Age Ranges"] = pd.cut(age_data["Age"], bins, labels=group_names)


age_demo_group = age_data.groupby("Age Ranges")

# Count players without repetitive names
age_count_demo = age_demo_group['SN'].nunique()
age_count_demo
```

```python
player_pct_demo = (age_count_demo/player_count)*100
player_pct_demo
```

```python
age_demo_df = pd.DataFrame({"Total Count":age_count_demo, "Percentage of Players":player_pct_demo})
age_demo_df
```
